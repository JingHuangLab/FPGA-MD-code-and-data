## Energy, Temperature, volume

```python
import numpy as np

import matplotlib.pyplot as plt

data_dir = '/home/chenjf/project/xiaojing/openmm-cmd/examples/07-DHFR-Explicit/'

x = np.linspace(0, 100, 50_000)
#"Step","Potential Energy (kJ/mole)","Kinetic Energy (kJ/mole)","Total Energy (kJ/mole)","Temperature (K)"
#  0                1                             2                          3                 4
```

```python
gpu_sp_log = np.loadtxt(data_dir + 'ther_gpu_single', delimiter=',', skiprows=1)
gpu_dp_log = np.loadtxt(data_dir + 'ther_gpu_double', delimiter=',', skiprows=1)
fpga_log = np.loadtxt(data_dir + 'langchao/ther_fpga', delimiter=',', skiprows=1)
```

## potential energy

```python
fig, ax = plt.subplots(figsize=(8,6), dpi=600)

ax.plot(x, gpu_sp_log[:,1], label = 'GPU sp', alpha=0.7)
ax.plot(x, gpu_dp_log[:,1], label = 'GPU dp', alpha=0.7)
ax.plot(x, fpga_log[:,1], label = 'FPGA', alpha=0.7)


ax.set_xlabel('Time (ps)', fontsize=28)
ax.set_ylabel('Potential energy (kJ/mol)', fontsize=28)
ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0), 
                    useMathText=True)
ax.yaxis.get_offset_text().set_fontsize(20)
ax.tick_params(axis='both', labelsize=24, size=10, width=2)
ax.legend(fontsize=16)

ax.set_xlim(0,100)

plt.show()
```

## Kinetic energy

```python
fig, ax = plt.subplots(figsize=(8,6), dpi=600)

ax.plot(x, gpu_sp_log[:,2], label = 'GPU sp', alpha=0.7)
ax.plot(x, gpu_dp_log[:,2], label = 'GPU dp', alpha=0.7)
ax.plot(x, fpga_log[:,2], label = 'FPGA', alpha=0.7)


ax.set_xlabel('Time (ps)', fontsize=28)
ax.set_ylabel('Kinetic energy (kJ/mol)', fontsize=28)
ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0), 
                    useMathText=True)
ax.yaxis.get_offset_text().set_fontsize(20)
ax.tick_params(axis='both', labelsize=24, size=10, width=2)
ax.legend(fontsize=16)

ax.set_xlim(0,100)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6), dpi=600)

ax.plot(x, gpu_sp_log[:,3], label = 'GPU sp', alpha=0.7)
ax.plot(x, gpu_dp_log[:,3], label = 'GPU dp', alpha=0.7)
ax.plot(x, fpga_log[:,3], label = 'FPGA', alpha=0.7)


ax.set_xlabel('Time (ps)', fontsize=28)
ax.set_ylabel('Total energy (kJ/mol)', fontsize=28)
ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0), 
                    useMathText=True)
ax.yaxis.get_offset_text().set_fontsize(20)
ax.tick_params(axis='both', labelsize=24, size=10, width=2)
ax.legend(fontsize=16)

ax.set_xlim(0,100)

plt.show()
```
